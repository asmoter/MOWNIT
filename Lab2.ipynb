### DataFrames

- Zaawansowany pakiet do działania na tabelkach nxm danych 
- Podobne do pandas DataFrames w Pythonie albo data.frame w R
- Specjalistyczne funkcje do statystyki
- należy zaistalować pakiety:

```julia
Pkg.add("DataFrames")
```

- DataFrame to rodzaj bazy danych in-memory
- Składa się z kolumn, do których odwołujemy się używając symboli
- Każda z kolumn może przechowywać inny typ (inaczej niż w dwuwymiarowych tablicach )
- od wersji 0.11 każda z kolumn jest typu Array{T,1} gdzie T jest okreslonym typem danych (np. Float64)
- strona pakietu: https://github.com/JuliaStats/DataFrames.jl
- dokumentacja najnowszej wersji 0.14 : https://juliadata.github.io/DataFrames.jl/stable/index.html

````julia
#include <stdio.h>
#include <gsl/gsl_blas.h>
#include <gsl/gsl_errno.h>
#include <gsl/gsl_matrix.h>
#include <gsl/gsl_vector.h>
#include <time.h>
#include <sys/times.h>
#include <unistd.h>
#include <sys/time.h>


double calculate_time(struct timeval start, struct timeval end){

    double Time = (end.tv_sec - start.tv_sec) * 1000.0;
    Time += (end.tv_usec - start.tv_usec)/1000.0;
    return Time;
}

double multiplicate_vectors(const gsl_vector *X, const gsl_vector *Y, double *result){

    return gsl_blas_ddot(X, Y, result);
}

void multiplicate_vector_by_matrix(gsl_matrix *A, gsl_vector *v, gsl_vector *RES){

    gsl_blas_dgemv(CblasNoTrans, 1.0, A, v, 0.0, RES);
}


int main() {

    struct timeval v_start, v_end;
    struct timeval m_start, m_end;
    int vector_size;
    gsl_vector *V1, *V2;
    double result;
    gsl_matrix *M;
    gsl_vector *RESULT;

    char* filePath = "/home/aleksandra/Dokumenty/Mownit/newResults.csv";
    FILE* results = fopen(filePath, "w+");

    if(!results){
        printf("Something went wrong with opening file");
    }

    fprintf(results, "Dim,Type,Time\n");

    for(int dim = 500; dim <= 5000; dim += 50){

        vector_size = dim;
        V1 = gsl_vector_alloc((const size_t) vector_size);
        V2 = gsl_vector_alloc((const size_t) vector_size);


        for(int j = 0; j < dim; j++){
            gsl_vector_set(V1, (size_t)j, rand()%130+j*15.2);
            gsl_vector_set(V2, (size_t)j, rand()%280+j*6.3);
        }

        M = gsl_matrix_alloc((const size_t) dim, (const size_t) dim);
        RESULT = gsl_vector_alloc((const size_t) dim);


        for(int i = 0; i < dim; i++){
            for(int j = 0; j < i; j++){
                gsl_matrix_set(M, (size_t)i, (size_t)j, rand()%70+i+j*3.5);
            }
        }

        char* type[20];
        double time[20];

        for(int j = 0; j < 20; j+=2){

            gettimeofday(&v_start, NULL);
            multiplicate_vectors(V1, V2, &result);
            gettimeofday(&v_end, NULL);

            type[j] = "v";
            double calculateTime_v = calculate_time(v_start, v_end);
            time[j] = calculateTime_v;

            gettimeofday(&m_start, NULL);
            multiplicate_vector_by_matrix(M, V1, RESULT);
            gettimeofday(&m_end, NULL);

            type[j+1] = "m";
            double calculateTime_m = calculate_time(m_start, m_end);
            time[j+1] = calculateTime_m;
        }

        for(int k = 0; k < 20; k++){
            fprintf(results, "%d,%s,%0.3f\n", vector_size, type[k], time[k]);
        }

        gsl_vector_free(V1);
        gsl_vector_free(V2);

        gsl_matrix_free(M);
        gsl_vector_free(RESULT);

    }

    return 0;
}
````
